```python
import matplotlib.pyplot as plt

with open('phipsi_output_file.txt', 'r') as file:
    lines = file.readlines()

phi_values = []
psi_values = []

for line in lines:
    line = line.strip('\n')
    values = line.split()

    phi = float(values[0])
    psi = float(values[1])

    phi_values.append(phi)
    psi_values.append(psi)

plt.scatter(phi_values, psi_values, s=5, c='b', alpha=0.5)
plt.xlabel('Phi')
plt.ylabel('Psi')
plt.title('Ramachandran Diagram')
plt.grid(True)
plt.show()
```
